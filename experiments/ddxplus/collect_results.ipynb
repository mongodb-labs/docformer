```python
import guild.ipy as guild

runs = guild.runs(labels=["ddxplus-11-eval-test"], completed=True)
scalars = runs.scalars()
scalars

# only keep rows where tag is one of ["ddp", "ddr", "f1", "gtpa", "gtpa_at_1"]
scalars = scalars[scalars["tag"].isin(["ddp", "ddr", "f1", "gtpa", "gtpa_at_1"])]

scalars = scalars.pivot(index="run", columns="tag", values="first_val")
scalars = scalars.sort_values(by="f1", ascending=False, ignore_index=True)

display(scalars)
results = scalars.aggregate(["mean", "std"])

results
```
